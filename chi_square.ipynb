```python
import nltk
nltk.download('popular')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
import pandas as pd
import numpy as np
```

<h1>Baca Dokumen CSV

```python
df = pd.read_excel('jurnal_terlabeli.xlsx')
contents = df['Abstrak_indo']
contents = contents.str.lower()
df.head()
```

<h1>Stemming Data

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stem_kata = []
jmlData, jmlKolom = df.shape 
for i in range(jmlData):
    # sentence = ' '. join([str(item) for item in contents[i]])
    output   = stemmer.stem(contents[i])
    stem_kata.append(output)
df['stem_kata'] = stem_kata
```

```python
df['stem_kata']
```

<h1>Stopword

```python
def stopping_word(contents):    
    data_kata = []
    stop_words = stopwords.words('english')
    stop_words2 = stopwords.words('indonesian')
    stop_words.extend(stop_words2)
    jmlData = contents.shape 
    for i in range(jmlData[0]):
        word_tokens = word_tokenize(contents[i])
        # print(word_tokens)
            
        word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]

        special_char = "+=`@_!#$%^&*()<>?/\|}{~:;.[],1234567890‘’'" + '"“”●'
        out_list = [''.join(x for x in string if not x in special_char) for string in word_tokens_no_stopwords]
        # print('List after removal of special characters:', out_list)

        while '' in out_list:
            out_list.remove('')
        data_kata.append(out_list)
    return data_kata
```

```python
stop_kata = stopping_word(df['stem_kata'])
df['stop_kata'] = stop_kata
```

```python
for i in range(jmlData):
    while '-' in df['stop_kata'][i]:
        df['stop_kata'][i].remove('-')
```

<h1>TF-IDF

```python
tfidf = TfidfTransformer(use_idf=True,norm='l2',smooth_idf=True)
tf=tfidf.fit_transform(vectorizer.fit_transform(df['stop_kata_join'])).toarray()
```

```python
df_Tf_Idf =pd.DataFrame(data=tf, columns=[a])
df_Tf_Idf.head(10)
```

```python
df_Tf_Idf_label = df_Tf_Idf
df_Tf_Idf_label["Kategori_biner"] = df["Kategori_biner"]
df_Tf_Idf_label.head(10)
```

```python
# df_Tf_Idf.to_csv('prepocessing_data_TF_IDF.csv', sep=',')
```

```python
a = df_Tf_Idf_label["Kategori_biner"].to_numpy()
a = a.T
a[0].T
```

<h1>Chi-Square

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.feature_selection import chi2, SelectKBest
```

```python
y_binarized = LabelBinarizer().fit_transform(a[0].T)
print(y_binarized)
```

```python
observed = np.dot(y_binarized.T, tf)
observed.shape
```

```python
class_prob = y_binarized.mean(axis = 0).reshape(1, -1)
feature_count = tf.sum(axis = 0).reshape(1, -1)
expected = np.dot(class_prob.T, feature_count)

print(expected)
```

```python
feature_count.shape
```

```python
chisq = (observed - expected) ** 2 / expected
chisq_score = chisq.sum(axis = 0)
print(chisq_score)
```

```python
chi2score = chi2(tf, a[0].T)
skorchi = np. reshape(chi2score[0], (1, 1395))
phisccore=np. reshape(chi2score[1], (1, 1395))
```
